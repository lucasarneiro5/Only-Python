##**1) *all()* function**

The all() function returns True if all items in an iterable are true, otherwise it returns False.

If the iterable object is empty, the all() function also returns True.

image.png

*0 = False


*1 = True

```
all(
    condition(item)
    fro item in iterable:
)
```



````python
# all values true
l = [1, 3, 4, 5]
print(all(l))
````

````python
mylist = [0, 1, 1]
x = all(mylist)
x
````

##**2) *any()* function**

The any() function returns True if any item in an iterable are true, otherwise it returns False.

If the iterable object is empty, the any() function will return False.

image.png

* 0 = False


* 1 = True

```
any(
    condition(item)
    fro item in iterable:
)
```



````python
mytuple = (0, 1, False)
x = any(mytuple)
x
````

````python
mydict = {0 : "Apple", 1 : "Orange"}
x = any(mydict)
x
````

## **Challenge:** Create a function that takes in an input string and returns true if the string contains a punctuation, and false if doesn't.
 

````python
import string
string.punctuation
````

````python
def test_with_punctuation():
    assert contains_punctuation('Readability counts') 
    assert contains_punctuation('If the implementation is hard to explain, it\'s a bad idea.')
    assert contains_punctuation('There should be one-- and preferably only one --obvious way to do it.')
````

````python
def contains_punctuation(input_str):
  '''Return True if the input_str contains punctuations.
     Return false otherwise'''
  return any(char in string.punctuation
             for char in input_str)
````

##**3) *enumerate()* function**

* Se necessita além do valores, das chaves (key) de cada valor.

image.png

https://realpython.com/python-enumerate/

````python
countries = ['Brazil', 'EUA', 'Nepal', 'Japan']

for index in range(len(countries)):
  print(f'{index+1}. {countries[index]}')
````

````python
# Using enumerate()
for item in enumerate(countries, start=1):
  print(item)
````

````python
for index, country in enumerate(countries, start=1):
  print(f'{index}.{country}')
````

##**4) *zip()* function**

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

image.png

````python
countries = ['Brazil', 'EUA', 'Nepal', 'Japan']
capitals = ['Brasilia', 'Washington', 'Kathmandu', 'Tokyo']
````

````python
for country, capital in zip(countries, capitals):
  print(f'The capital city of {country} is {capital}')
````

````python
pairs = list(zip(countries, capitals))
pairs
````

````python
country, capital = zip(*pairs)
````

## **Challenge:**


````python
years = [2010,2011,2012,2013,2014,2015]
births=[723_165,723_913,729_674,698_512,695_512,697_852]
````

````python
def annual_births_average(year=years, births=births):
  '''Return a list of tuples with each entry in this format
    (year, birth, runnig_avarage)
    Round the running_average to the nearest integrer.
  '''
  result = []
  sum_ = 0

  for index, (year, birth) in enumerate(zip(years,births), start=1):
    sum_ += birth
    result.append((year, birth, round(sum_ / index)))

  return result
````
