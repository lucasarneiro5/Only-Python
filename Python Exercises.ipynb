### N.32) Read a phrase and if it is a 'plindromo'. A phrase that can be read reversing it.

```python
phrase = str(input('Write a phrase: ')).upper().strip()
word = phrase.split() # Separate each word
t = ''.join(word) # Join each word without space
print('You wrote: {}'.format(t))
inverse = ''

for l in range(len(t) - 1, -1, -1): # The size - 1 to the last letter, until the first letter, decreasing -1
    inverse += t[l] # Bulding a list
if inverse == t:
    print('\n\nThis a palindromo! {} = {}.'.format(t, inverse))
else:
    print('\n\nThis is not a palindromo!')
    
#or
# inverse = t[::-1] replacing 'for'. From begining(:), to the end (:), decreasing
```

### *N.43) Build a cash machine that works with 50, 20, 10 and 1 bank notes:

```python
print('='*30)
print('{:^30}'.format('PYTHON BANK'))
print('='*30)
value = int(input('How much do you want? $'))
total = value
c = 50 # $50 bank note
bn = 0 # Bank notes total

while True:
    if total >= c:
        total -= c # Take $50 from value
        bn += 1 # It'll check how many $50 you can take
    else:
        if bn > 0:
            print('You need {} bank notes of ${}'.format(bn, c))
        if c == 50:
            c = 20
        elif c == 20:
            c = 10
        elif c == 10:
            c = 1
        bn = 0
        if total == 0:
            break
print('='*30)
print('{:^30}'.format('COME BACK SOON!'))
```

### N.64) Calculate the factorial of a number

```python
def factorial(n, show=False):
    """
    -> Calculate a factorial of a number.
    
    """
    f = 1
    for c in range(n, 0, -1):
        if show:
            print(c, end='')
            if c > 1:
                print(' x ', end='')
            else:
                print(' = ', end='')
        f *= c
    return f

print(factorial(5, show=True)) #Show the result
#help(factorial)
```
