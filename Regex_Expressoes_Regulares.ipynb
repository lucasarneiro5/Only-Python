```python
# Regex library
import re

re.compile('Lucas')
```

```python
# Search d√° a posicao que estar√° a string procurada
re.search('a', 'Lucas')
```

```python
re.search('a', 'Arara')
```

```python
# Match busca somente no inicio
re.match('a', "Lucas")
```

```python
#Split
'Lucas'.split('a')
```

```python
# Split do Regex
re.split(r'a', 'Lucas')
```

```python
re.split(r'(c|a)', 'Lucas')
```

```python
# Findall = Mostra todos os 'a'
re.findall('a|A', 'Arara')
```
